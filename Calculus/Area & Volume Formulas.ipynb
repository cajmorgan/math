# Area & Volume Formulas
## Deriving formulas using Calculus

Many formulas for area and volume we learn as childrens or teenagers can be derived directly from calculus using integrals. The reason for this is of course pretty self-explanatory, but calculus gives us the tool to basically calculate the area and volume of stuff as long as we have a function to derive it from. The elegant part of this approach is that we can actually find the **exact** formulas by first imagining how integrate an object and it's extremely beautiful. This will be a collection of some of those formulas.


## Area

### Circle - $\pi r^2$
The area of a circle is probably the first formula for area that might not be so easy to explain. Why is it $r^2$ and how can this become the area? There are multiply ways to explain this formula, some easier than others. Here I'm going to derive it step by step using integration.

<img src="./pngs/circle-area.png" width=400>

This image is a good start to solve this problem intuitively, imagine we have this triangle where the $y$ side has a thickness of $\Delta x$. This triangle moves from $-r$ to $r$ (left to right) and collects the whole area step by step. In this case we have drawn the $y$ - line only covering half of the circle. 

We want to integrate this with the change of $x$, therefore we need to find a formula for $y$. Using the radius:

$r^2 = x^2 + y^2$

$r^2 - x^2 = y^2$

$\sqrt{r^2 - x^2} = y$

where 
$-r \leq x \leq r$  

Good, now we have the function we need to integrate this. 
To get the whole area, we multiply the whole integral by $2$.

### $ A = 2\int_{-r}^{r} \sqrt{r^2 - x^2} \,dx$

And we could actually end it here as we have the integral and we can plug it into a calculator and you will see it gives an exact answer to the area of a circle (as exact it can be) using integration, but that wouldn't be so fun, right? This is also a good time to just think a bit about integration as a whole with a infinitesimal perspective, we swipe this line with a change of $\Delta x$ and sum up the area $y \Delta x$ for every $\Delta x$ that can "fit" the circle resulting in the total area. This can also be written as 

### $2 \sum_{i = 1}^{n} 2\sqrt{r^2 - (\Delta xi)^2} \; \Delta x $ where $\Delta x = \frac{r}{n}$

Riemann sums are explained further under the "integral" chapter. 

So far so good, but how do we actually derive the formula from that integral manually?

As we can't use the inverse chain rule because we have a $x^2$ inside the square root, we can actually solve this by substitution using trignometric functions, which I must admit, is kinda fitting as we are working with the area of a circle. 

### $ A = 2\int_{-r}^{r} \sqrt{r^2 - x^2} \,dx$

$x = rsin(\theta)$

$\theta = sin^{-1}(\frac{x}{r})$

$\frac{dx}{d\theta} = rcos(\theta) \times 1$

$dx = rcos(\theta) \times d\theta$ | Later we need to substitute dx as we are deriving with respect to $\theta$ instead of $x$



### $ A = 2\int_{-r}^{r} \sqrt{r^2 - r^2sin^2\theta} \,dx =$

### $ 2\int_{-r}^{r} \sqrt{r^2 - r^2sin^2\theta} \; rcos(\theta)\,d\theta = $

### $ 2\int_{-r}^{r} \sqrt{r^2( 1 - sin^2\theta}) \; rcos(\theta)\,d\theta = $

### $ 2\int_{-r}^{r} \sqrt{r^2(cos^2\theta}) \; rcos(\theta)\,d\theta = $

### $ 2\int_{-r}^{r} r^2cos^2\theta \; d\theta = $

### $cos^2\theta = \frac{cos(2\theta) + 1}{2} = $

### $ 2r^2\int_{-r}^{r} \frac{cos(2\theta) + 1}{2} \; d\theta$

### $ F(x) = 2r^2  [\frac{sin(2\theta)}{4} + \frac{\theta}{2}] $
The above function is our anti-deriative function. Let's plug in $r$ and $-r$ and see where we get

### $ F(r) = 2r^2  [\frac{sin(2(sin^{-1}(\frac{r}{r})))}{4} + \frac{sin^{-1}(\frac{r}{r})}{2}] = $

### $ F(r) = 2r^2  [0 + \frac{\pi}{4}]$

### $ F(-r) = 2r^2  [0 - \frac{\pi}{4}]$

### $ F(r) - F(-r) = 2r^2  [\frac{\pi}{4}] + 2r^2  [\frac{\pi}{4}] = $

## $\frac{4r^2\pi}{4} = r^2\pi $



## Cylinder $\pi r^2 h$

### Volume

As long as we know the area of a circle, the volume of a cylinder is pretty straight forward. Let's imagine that we slide a circle towards $h$ and sum the area for every $dy$

### $\int_0^{h} r^2\pi dy$


The antiderivative function simply becomes the formula

$F(h) = \pi r^2 h$

## Cone $\pi\frac{r^2h}{3}$

### Volume

The volume of a cylinder is pretty intiutive, yea of course the area of a circle times the height produces the volume, it's obvious! With a cone, it's actually NOT so obvious, but it has a beautiful answer. What's is really fine with this answer is that it works for any cone regardless of the relationship between the height and the radius. 

First we need to find a general line that we can use, which is the slope of the cone:

$y = h - kx : 0 \leq kx \leq h$ 

This line tells us, when $kx = 0$ we are at the maximum point $h$ of the cone, when $kx = h$ we are at the lowest point $0$. 

We could integrate this with respect to $x$ but for a more realistic "standing" cone, we want to integrate this with respect to $y$. This mean that we can imagine the line creating the volume of the cone while rotating around the y-axis.

$y = h - kx$

$x = \frac{h - y}{k}$


Let's integrate the above function $x$ with respect to $y$.

### $\displaystyle\int_0^{h} \pi x^2 dy = $

### $\displaystyle\int_0^{h} \pi (\frac{h - y}{k})^2 dy $

Let's find the anti-derivative function using the chain rule

### $F(y) = \frac{-k\pi(\frac{h - y}{k})^3}{3} $

Plug in $h$ and $0$

### $F(h) = \frac{-k\pi(\frac{h - h}{k})^3}{3} = 0$

### $F(0) = \frac{-k\pi(\frac{h}{k})^3}{3} =  $

### $F(0) = \frac{-\pi\frac{h^3}{k^2}}{3}$

So we are done... But it doesn't look like the formula?
This is the beautiful moment, let's think a moment, what is $k$ really? 
Well, $k$ is the slope between 2 points right?

### $\frac{y_1 - y}{x_1 - x} = k$

The points in every cone that we are looking for is of course, the maximum $y$ height $h$ and the maximum $x$ radius $r$, these points will be $p_1 = (0, h), \; p_2 = (r, 0)$. 

### $\frac{h - 0}{0 - r} = k = $
### $-\frac{h}{r} = k$

Substitute $k$ which will turn positive thanks to the square

### $F(0) = \frac{-\pi\frac{h^3}{\frac{h^2}{r^2}}}{3} = $

### $F(0) = \frac{-\pi\frac{r^2h^3}{h^2}}{3} = $

### $F(0) = -\pi\frac{r^2h}{3}$

### $F(h) - F(0) = 0 + \pi\frac{r^2h}{3}$

## $\pi\frac{r^2h}{3}$

Perfection. The part where $k$ lines up and cancels everything perfectly for every cone is my favorite. Even after deriving this formula, it still feels a bit like magic, the number $\frac{1}{3}$, why that number specifically? Of course we know it's because of the integration, but still, it's just so cool to arrive from such a small and simple answer.

Learning the sphere volume formula as a kid was probably the most memorable weird moment for me in math, why the hell are we doing $\frac{4}{3}r^3\pi$? I wasn't particularly interested in math back then but I think I asked my teacher this because it didn't make sense, and I didn't get any particular good response. Learning Calculus as an adult gave me the answer and it was more beautiful and simple than I could ever imagine.

Imagine a sphere where you swipe a circle through and sum the area of that circle.

<img src="pngs/sphere-volume.png" width=300 />

The equation of the circle is:

$r^2 = x^2 + y^2$

To get y, we do the following:

$y = \sqrt{r^2 - x^2}$

The reason why we need $y$ is because when we are at point $x$ in a circle, $y$ won't be the same, they are always working in opposite directions, so instead of working with two variables, we will only work with one:

$A(x) = \pi y^2$

$A(x) = \pi(r^2 - x^2)$

At any given x, this will yield the area of a circle at that particular point going through $y$ and slicing the whole sphere, similarily to the picture above. Let's integrate:

$\displaystyle\int_{-r}^{r} \pi(r^2 - x^2)dx $

#### $F(x) = \pi r^2x - \pi\frac{x^3}{3}$

### $F(x) = \frac{3\pi r^2x - \pi x^3}{3}$

### $F(r) - F(-r) = \frac{3\pi r^3 - \pi r^3}{3} + \frac{3\pi r^3 - \pi r^3}{3} = $

### $F(r) - F(-r) = \frac{2\pi r^3}{3} + \frac{2\pi r^3}{3} = $

### $F(r) - F(-r) = \frac{4\pi r^3}{3} $
