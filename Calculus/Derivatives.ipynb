#### Chain Rule
When using the chain rule, as an example it's good to consider a function that gets an effect from another function, just like a chain of events. F.e let's say we have the function $V(r)$ that depends on the variable $r$, the radius. Let's say that $r$ depends on time in seconds $r(t)$. The most important thing here to understand is that the variable or function $r$ is dependent or independent completely depending on **context**. 

What would the derivative of $V(r)$ be? Well, it will just be $\frac{dV}{dr}$, simple as usual, but what would the derivative of $V(t) = V(r(t))$ be? $V(t)$ first depends on $r$ which depends on $t$, this means that we can't use derive it as usual, here we need the chain rule, which states:

$\frac{dV}{dt} = \frac{dV}{dr} \times \frac{dr}{dt}$

The nice thing about the chain rule is that the ratio (which some say you shouldn't treat as a ratio) cancels out nicely.

## Implicit and Explicit Derivatives

### Implicit Derivative
When deriving implicitly, f.e $y^2 = x^2 + 1$ we can derive this with respect to $x$, but as we know $y$ is a function of the input $x$ we need to use the **Chain Rule** everytime we take the derivative of $y$ with respect to $x$. **"Respect to x"** simply means that we are applying the change to $x$ f.e $\frac{f(x + h) - f(x)}{h}$ where h is the change and therefore $y$ is the resulting effect of applying that change to $x$. This is the reason in implicit derivatives, we can derive $x$ explicitly but not $y$.

In the following example, $y^2$ is actually a set of two functions $f(g(x))$ as y itself is a function of x and then we have the outer function exponent $f(y) = y^2$ which means what we are really taking the derivative of on the left side is both the inner function of $y = g(x)$ and the outer $f(y) = y^2$. Easy example:

$y^2 = x^2 + 1$
