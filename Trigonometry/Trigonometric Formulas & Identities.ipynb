## Laws of Trigonometric Functions



### Law of Sine

Formula: $$\frac{sin(A)}{a} = \frac{sin(B)}{b} = \frac{sin(C)}{c}$$ 


#### Proof
This is a general algebraic/visual proof of Law of Sines which is pretty straight forward and easy to understand.

<img src="./svgs/law-of-sine.svg" width=600 align="center"> 


$\sin(A) = \frac{h_1}{a}$

$\sin(B) = \frac{h_1}{b}$

$\sin(C) = \frac{h_2}{a}$

$\sin(A) = \frac{h_2}{c}$


$a \times sin(B) = b \times sin(A)$
$a \times sin(C) = c \times sin(A)$


$\frac{sin(A)}{a} = \frac{sin(B)}{b} = \frac{sin(C)}{c}$

### Law of Cosine

Formula: $$a^2 = \sqrt{b^2 + c^2 - 2bc\times\cos(A)} $$
    
#### Proof
This is a general algebraic/visual proof of Law of Cosines which is a bit more complex than the previous. In this case we are using a triangle with an obtuse angle. The proof would work a little bit different for a triangle with only acute angles, but, the same logic applies.

We first draw an extra "ghost" triangle (in yellow) to the left and then solve for $a$. Note that $\cos(180 - A) = -\cos(A)$

<img src="./svgs/law-of-cosine.svg" width=600 align="center"> 

$a = \sqrt{(c_1+c)^2 + b_1^2}$

$b_1 = \sqrt{b^2 - c_1^2}$

$\frac{c_1}{b} = -\cos(A)$

$c_1 = -\cos(A)\times b$

Substitute $c_1$ and $b_1$ in $a$, notice the exponent in $b_1$:

$a = \sqrt{(-\cos(A)\times b + c)^2 + b^2 - c_1^2}$

$a = \sqrt{\cos^2(A)\times b^2 + c^2 - 2ab\cos(A) + b^2 - \cos^2(A)\times b^2}$

$a = \sqrt{b^2 + c^2 - 2ab\cos(A)}$

*Note that some substitution steps are skipped here, but if you solve the equation you will see them*

### Square formulas
Square formulas are very handy, especially when applying calculus to trigonometric functions as they can be surprisingly hard to integrate in their corresponding squared form. 

## $cos^2(v) = \frac{1 + cos(2v)}{2}$

Getting $cos^2(v)$ in a non-fractional form is way easier to derive and integrate than in a squared form. By combining the double angle formula and identity we can arrive at a handy result. 

$cos^2(v) - sin^2(v) = cos(2v)$

$cos^2(v) - cos(2v) = sin^2(v)$

$1 - cos^2v = sin^2v$

$cos^2(v) - cos(2v) = 1 - cos^2(v)$

$2cos^2(v) = 1 + cos(2v)$

$cos^2(v) = \frac{1 + cos(2v)}{2}$

### Product-Sum formulas

Might be used rarely, but can be handy in very specific situations solving trigonometric equations by hand.

## $sin(a)sin(b) = \frac{cos(a - b) - cos(a + b)}{2}$

$cos(a)cos(b) = cos(a + b) + sin(a)sin(b)$

$cos(a)cos(b) = cos(a - b) - sin(a)sin(b)$

$cos(a + b) + sin(a)sin(b) = cos(a - b) - sin(a)sin(b)$

$2sin(a)sin(b) = cos(a - b) - cos(a + b)$

$sin(a)sin(b) = \frac{cos(a - b) - cos(a + b)}{2}$


## $cos(a)cos(b) = \frac{cos(a - b) + cos(a + b)}{2}$

$sin(a)sin(b) = cos(a)cos(b) - cos(a + b)$

$sin(a)sin(b) = cos(a - b) - cos(a)cos(b)$

$cos(a)cos(b) - cos(a + b) = cos(a - b) - cos(a)cos(b)$

$cos(a)cos(b) = \frac{cos(a - b) + cos(a + b)}{2}$


## $cos(a)sin(b) = \frac{sin(a+b) - sin(a-b)}{2}$

From addition formulas: 

$sin(a + b) -  cos(a)sin(b) = sin(a)cos(b)$

$sin(a - b) + cos(a)sin(b) = sin(a)cos(b))$

$2cos(a)sin(b) = sin(a + b) - sin(a - b)$

$cos(a)sin(b) = \frac{sin(a+b) - sin(a-b)}{2}$

## $sin(a)cos(b) = \frac{sin(a + b) + sin(a - b)}{2}$

$cos(a)sin(b) = sin(a + b) - sin(a)cos(b)$

$cos(a)sin(b) = sin(a)cos(b) - sin(a - b)$

$sin(a + b) - sin(a)cos(b) = sin(a)cos(b) - sin(a - b)$

$2sin(a)cos(b) = sin(a + b) + sin(a - b)$

$sin(a)cos(b) = \frac{sin(a + b) + sin(a - b)}{2}$


### Sum-Product formulas
To derive the sum-product formulas, we can use the formulas above.

Before prooving the formulas, let's declare som general variables:

$x = a + b, y = a - b$

$x + y = 2a$

$\frac{x + y}{2} = a$

$\frac{x - y}{2} = b$

As this formula will be perfect to use when we DON'T know $a, b$, therefore, let's use it in the context of $x, y$.

## $sin(x) + sin(y) = 2sin(\frac{x + y}{2})cos(\frac{x - y}{2})$

$sin(x) + sin(y) = sin(a + b) + sin(a - b) = 2sin(a)cos(b)$

$sin(x) + sin(y) = 2sin(\frac{x + y}{2})cos(\frac{x - y}{2})$


## $sin(x) - sin(y) = 2cos(\frac{x + y}{2})sin(\frac{x - y}{2})$

$sin(x) - sin(y) = sin(a+b) - sin(a-b) = 2cos(a)sin(b)$

$sin(x) - sin(y) = 2cos(\frac{x + y}{2})sin(\frac{x - y}{2})$

## $cos(x) + cos(y) = 2cos(\frac{x + y}{2})cos(\frac{x - y}{2})$

$cos(x) + cos(y) = cos(a + b) + cos(a - b) = 2cos(a)cos(b)$

$cos(x) + cos(y) = 2cos(\frac{x + y}{2})cos(\frac{x - y}{2})$



## $cos(x) - cos(y) = -2sin(\frac{x + y}{2})sin(\frac{x - y}{2})$

$cos(x) - cos(y) = cos(a + b) - cos(a - b) = -2sin(a)sin(b)$

$cos(x) - cos(y) = -2sin(\frac{x + y}{2})sin(\frac{x - y}{2})$


#### Handling summing of different functions

## $sin(x) + cos(y) = 2sin(\frac{x - y}{2} + 45)cos(\frac{x + y}{2} - 45)$

## $sin(x) - cos(y) = 2cos(\frac{x - y}{2} + 45)sin(\frac{x + y}{2} - 45)$

## $cos(x) + sin(y) = 2cos(\frac{x - y}{2} + 45)cos(\frac{x + y}{2} - 45)$

## $cos(x) - sin(y) = -2sin(\frac{x - y}{2} + 45)sin(\frac{x + y}{2} - 45)$



In order to convert $sin(x) + cos(y)$ into a product, we can change a couple of things:

$x = a + b, y = 90 -a + b$

$x + y = 90 + 2b$

$x - y = 2a - 90$

$\frac{x - y}{2} + 45 = a$

$\frac{x + y}{2} - 45= b$

We do this because, $sin(x) + sin(y) = sin(x) + cos(90 - y)$

Following the formula for $sin(x) + sin(y)$, with this new y, we would get cos(y) instead. Just replace $a, b$

$sin(x) + cos(y) = 2sin(a)cos(b)$

$sin(x) + cos(y) = 2sin(\frac{x - y}{2} + 45)cos(\frac{x + y}{2} - 45)$
