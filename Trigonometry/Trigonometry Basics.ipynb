### Degrees & Radians
Radians and degrees are the exact same thing, but in different forms. Radians is the natural version of it and degrees are the human invention (notice that 360 is very close to a full year) that makes it more acessible in certain scenarios. Probably the most important angle in mathematics are perpendicular, $90^\circ$ which is the same thing as $\frac{\pi}{2}$. 

Radians are defined as the same length as the radius of a unit circle, 1. It's perfectly logical as that means we have $2\pi$ radians for every circle regardless its size. $360^\circ = 2\pi$. Degrees are handy in certain situations, like classic geometry problems, but what about radians? Radians come up all over the place, especially in analysis. Something important to consider is that degrees never comes up natural in formulas in mathematics.
#### Degrees -> Radians
1 degree in radians is expressed as: 

$1^\circ = \frac{2\pi}{360^\circ } = \frac{\pi}{180^\circ } = 0.017453292519943 $

And to get the radians we multiply by the amount of degrees $d$. 

$\frac{\pi}{180^\circ } * d$

Example, how many radians are $45^\circ$?

$\frac{45^\circ \pi}{180^\circ } = \frac{\pi}{4}$



#### Radians -> Degrees
To convert from radians to degrees, we do the reverse:

1 radian in degrees:

$1r = \frac{180^\circ }{pi} =  57,295779513082321^\circ$

To get the degrees we multiply by the amount of radians $r$.

$\frac{180^\circ }{pi} * r$

Example, how many degrees are $\frac{\pi}{4}$

$\frac{180^\circ }{pi} * \frac{\pi}{4} = \frac{180^\circ }{4} = 45^\circ$

### Using Trigonometry to prove $\angle\alpha+\angle\beta+\angle\gamma = 180^{\circ}$ 
This is one Trigonometry proof for prooving that the sum of the angles in a triangle is always $180^{\circ}$. This proof can be directly written out from a triangle drawing through reasoning, but we will do it through algebra.

$\alpha + \beta + \gamma = 180$ 

$\sin(\alpha + \beta + \gamma) = 0$ 

To solve it in this direction, the addition formula for sine is handy (proven in other document):

$\sin(x + y) = \sin(x)\times\cos(y) + \sin(y)\times\cos(x)$


#### Derivation

$\sin(\alpha + (\beta + \gamma)) = 0$ 

$\sin(\alpha)\times\cos(\beta+\gamma) + \sin(\beta+\gamma)\times\cos(\alpha) = 0 $

$\sin(\alpha)\times\cos(\beta+\gamma) = -\sin(\beta+\gamma)\times\cos(\alpha) $

$\frac{\sin(\alpha)}{\cos(\alpha)} = \frac{\sin(\beta+\gamma)}{-\cos(\beta+\gamma)} $

This means that, when $\sin(\alpha + (\beta + \gamma)) = 0$ 

$\sin(\alpha) = \sin(\beta+\gamma)$ & $\cos(\alpha) = -\cos(\beta+\gamma)$


Note that we can continue using the addition formula to solve for $\beta$ & $\gamma$ with the same respective results.
We can aquire:

$\frac{\sin(\beta)}{\cos(\beta)} = \frac{\sin(\alpha+\gamma)}{-\cos(\alpha+\gamma)} $
Which we want for the following drawing.

This might be pretty obvious if we just visualize a circle and some basics trigonometry rules, but the point here is to visualize this with a triangle and how that can prove the statement.
